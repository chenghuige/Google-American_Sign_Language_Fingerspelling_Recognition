```python
frames = df['frames'].values[0]
```

```python
def all_nan(frame):
  return np.isnan(frame).all()
```

```python
def hand_nan(frame):
  return np.isnan(frame[:84]).all()
```

```python
def left_hand_nan(frame):
  return np.isnan(frame[:42]).all()
```

```python
def right_hand_nan(frame):
  return np.isnan(frame[42:84]).all()
```

```python
# pdl.initialize(nb_workers=4, progress_bar=True)
```

```python
df['n_all_nan_frames'] = df['frames'].progress_apply(lambda frames: sum(all_nan(x) for x in frames))
```

```python
df['n_hand_nan_frames'] = df['frames'].progress_apply(lambda frames: sum(hand_nan(x) for x in frames))
```

```python
df['n_lhand_nan_frames'] = df['frames'].progress_apply(lambda frames: sum(left_hand_nan(x) for x in frames))
```

```python
df['n_rhand_nan_frames'] = df['frames'].progress_apply(lambda frames: sum(right_hand_nan(x) for x in frames))
```

```python
cols = ['n_frames', 'n_frames_per_char', 'n_all_nan_frames', 'n_hand_nan_frames', 'n_lhand_nan_frames', 'n_rhand_nan_frames']
d = df[cols]
d
```

```python
d.describe()
```

```python
df['n_hand_frames'] = df['n_frames'] - df['n_hand_nan_frames']
df['n_hand_frames_per_char'] = df['n_hand_frames'] / df['phrase_len']
df['hand_frames_ratio'] = df['n_hand_frames'] / df['n_frames']
cols += ['n_hand_frames', 'n_hand_frames_per_char', 'hand_frames_ratio']
d = df[cols]
d
```
